# Data Cleaning of Each Dataset

I have decided to not merge the datasets, as they are separate, and the anime_user_df data set is dependent on the other two. Once I have cleaned each dataset, by handling all of the missing values I will load the clean datasets into a SQLLite database.

# A) User List

```python
user_df.info()
```

```python
user_df.columns
```

```python
user_df.dtypes
```

```python
user_df.isnull().sum()
```

```python
user_df.drop(['gender', 'location', 'birth_date', 'access_rank'], axis=1, inplace=True)
```

```python
user_df.isnull().sum()
```

```python
user_df.head()
```

```python
user_df['join_date'] = user_df['join_date'].ffill()
```

```python
user_df['last_online'] = user_df['last_online'].ffill()
```

```python
user_df.isnull().sum()
```

```python
user_df['stats_mean_score'].fillna(user_df['stats_mean_score'].mean(), inplace=True)
user_df['stats_rewatched'].fillna(user_df['stats_rewatched'].mean(), inplace=True)
user_df['stats_episodes'].fillna(user_df['stats_episodes'].mean(), inplace=True)
```

```python
user_df.dropna(subset=['username'], inplace=True)
```

```python
user_df.isnull().sum()
```

```python
user_df.head()
```

# B) Anime List

```python
anime_df.info()
```

```python
anime_df.columns
```

```python
anime_df.isnull().sum()
```

```python
anime_df.drop(['title_english', 'title_japanese', 'title_synonyms', 'image_url', 'background', 'premiered', 'broadcast', 'producer', 'licensor'], axis=1, inplace=True)
```

```python
anime_df.isnull().sum()
```

```python
anime_df['rank'].fillna(value=anime_df['rank'].max() + 1, inplace=True)
anime_df['studio'].fillna(value='Unknown', inplace=True)
anime_df['genre'].fillna(value='Unknown', inplace=True)
```

```python
anime_df.head()
```

# C) UserAnime List

```python
anime_user_df.head()
```

```python
anime_user_df.info()
```

```python
anime_user_df.isnull().sum()
```

```python
anime_user_df.drop(['my_rewatching', 'my_tags'], axis=1, inplace=True)
anime_user_df.dropna(subset=['username'], inplace=True)
```

```python
animes_in_anime_list = anime_user_df['anime_id'].isin(anime_df['anime_id'])
```

```python
missing_animes = ~animes_in_anime_list.any()
```

```python
users_in_user_list = anime_user_df['username'].isin(user_df['username'])
missing_users = ~users_in_user_list.any()
```

```python
print(f"Missing animes: {missing_animes}, Missing users: {missing_users}")
```

I have cleaned each dataset removing any unnecessary columns and filling any missing value respectively. I have also checked the UserAnimeList to make sure that the values match to it's respective dataset as it is dependent. Now I will load into a SQL Lite Database.

```python
import sqlite3
```

```python
conn = sqlite3.connect('anime_data.db')
```

```python
user_df.to_sql('UserList', conn, if_exists='replace', index=False)
```

```python
anime_df.to_sql('AnimeList', conn, if_exists='replace', index=False)
```

```python
chunk_size = 10000
num_chunks = len(anime_user_df) // chunk_size + 1

for i in range(num_chunks):
    print(f'Saving chunk {i+1}/{num_chunks}')
    anime_user_df[i*chunk_size:(i+1)*chunk_size].to_sql('AnimeUserList', conn, if_exists='append', index=False)

```

```python
conn.close()
```
